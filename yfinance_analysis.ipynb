```python
import yfinance as yf
import pandas as pd
```

```python
tickers=["AAPL","MSFT","GOOGl","AMZN"]

#Extracting Data
raw_data=[]
for t in tickers:
    company=yf.Ticker(t)
    info=company.info

    raw_data.append({"Ticker":t,
    "Company_Name":info.get("longName"),
    "Sector":info.get("sector"),
    "Market_Cap":info.get("marketCap"),
    "PE_Ratio":info.get("trailingPE"),
    "Revenue_Growth":info.get("revenueGrowth"),
    "Profit_Margin":info.get("profitMargins"),
    "Operating_margin":info.get("operatingMargins"),
    "ROE":info.get("returnOnEquity"),
    "Current_Ratio":info.get("currentRatio"),
    "Debt_to_Equity":info.get("debtToEquity"),
    "Rev_Per_Employee":info.get("revenuePerEmployee"),
    "Total_Revenue":info.get("totalRevenue"),
    "Employee_Count":info.get("fullTimeEmployees")})
```

```python
#Data Frame Creation

df=pd.DataFrame(raw_data)
```

```python
#Counting Revenue Per Employee Manually as it has Null Value

df['Rev_Per_Employee']=df['Total_Revenue']/df['Employee_Count']
```

```python
#Converting Units for Easier Understanding
df['Market_Cap_Billions']=df['Market_Cap']/1_000_000_000
df['Total_Rev_Billions']=df['Total_Revenue']/1_000_000_000
df['Rev_Per_Employee_K']=df['Rev_Per_Employee']/1_000
df['Employee_Count']=df['Employee_Count'].fillna(0).astype(int)
```

```python
df=df.round({'Market_Cap_Billions':2,'Total_Rev_Billions':2,'Rev_Per_Employee_K':2,'Profit_Margin':4,'PE_Ratio':2,'Revenue_Growth':4})
```

```python
df=df.drop(columns=['Market_Cap','Total_Revenue','Rev_Per_Employee']) 
```
